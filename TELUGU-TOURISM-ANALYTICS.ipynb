2) List down the top 3 districts based on compound annual growth rate (CAGR) of visitors between (2016-2019)

```python
df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y')

# Create a year column
df['year'] = df['date'].dt.year

# Group the data by district and year, and sum the visitors
grouped = df.groupby(['district', 'year'])['visitors'].sum().reset_index()

# Pivot the table to have years as columns
cagr_df = grouped.pivot(index='district', columns='year', values='visitors').reset_index()

# Fill any NaN values with 0
cagr_df.fillna(0, inplace=True)

# Rename the columns
cagr_df.columns = ['district', '2016', '2017', '2018', '2019']

# Print the resulting dataframe
print(cagr_df)
```

```python
df
#cagr_df.dtypes
```

```python
cagr_df=cagr_df.assign(Compound_Annual_Growth_Rate = lambda x: (((x["2019"]/x["2016"])**0.25)-1))
cagr_df
```

```python
cagr_df.fillna(0)
```

```python
cagr_df.replace([np.inf, -np.inf], 2, inplace=True)
cagr_df
```
