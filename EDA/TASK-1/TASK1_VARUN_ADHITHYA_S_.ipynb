```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'],df['Monthly Income'])
plt.xlabel("AGE")
plt.ylabel("INCOME")
plt.title('Relationship between Age and Salary')
plt.show()
```

We can see that as the age increases the income increases Now let's check for linearity, exponentiality,logistic function as it looks like that

```python
from scipy.optimize import curve_fit

# Assuming 'x' and 'y' are the numerical variables from the plotted data

# Define the functions for linear, exponential, and logistic curves

def linear_func(x, a, b):
    return a * x + b

def exponential_func(x, a, b, c):
    return a * np.exp(b * x) + c

def logistic_func(x, a, b, c):
    return c / (1 + np.exp(-a * (x - b)))

# Fit the linear model
popt_linear, pcov_linear = curve_fit(linear_func, x, y)

# Fit the exponential model
popt_exponential, pcov_exponential = curve_fit(exponential_func, x, y)

# Fit the logistic model
popt_logistic, pcov_logistic = curve_fit(logistic_func, x, y)

# Calculate the R-squared values for each model
ss_total = np.sum((y - np.mean(y)) ** 2)
ss_residual_linear = np.sum((y - linear_func(x, *popt_linear)) ** 2)
ss_residual_exponential = np.sum((y - exponential_func(x, *popt_exponential)) ** 2)
ss_residual_logistic = np.sum((y - logistic_func(x, *popt_logistic)) ** 2)

r2_linear = 1 - (ss_residual_linear / ss_total)
r2_exponential = 1 - (ss_residual_exponential / ss_total)
r2_logistic = 1 - (ss_residual_logistic / ss_total)

# Print the R-squared values
print(f"R-squared (Linear): {r2_linear}")
print(f"R-squared (Exponential): {r2_exponential}")
print(f"R-squared (Logistic): {r2_logistic}")
```

This shows a linear relationship is better as it has the highest R-squared value which explains 63.53% of the data in the model

```python
x = df['Age']
y = df['Monthly Income']

# Calculate the regression line
m, b = np.polyfit(x, y, 1)  # Linear regression: m = slope, b = intercept
fit_line = m * x + b

# Plot the scatter plot and fit line
plt.figure(figsize=(10, 8))
plt.scatter(x, y)
plt.plot(x, fit_line, color='red')  # Plot the fit line
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Relationship between Age and Salary with Fit Line')
plt.show()
```
